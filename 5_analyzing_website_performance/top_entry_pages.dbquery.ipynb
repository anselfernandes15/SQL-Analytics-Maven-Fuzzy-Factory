```sql
-- top pages where the customer lands on your page for the first time 

CREATE OR REPLACE TEMP VIEW first_page_view AS
select 
  website_session_id,
  min(website_pageview_id) as min_pv_id,
  pageview_url
from workspace.mavenfuzzyfactory.website_pageviews 
where created_at < '2012-06-12'
group by 1,3;

select 
  pageview_url, 
  count(distinct website_session_id) as sessions
from first_page_view
group by 1
order by 2 desc;

select 
  fpv.pageview_url,
  count(distinct fpv.website_session_id) as sessions
from first_page_view fpv
join mavenfuzzyfactory.website_pageviews wps
on fpv.min_pv_id = wps.website_pageview_id
group by 1
```
