```sql
select 
  ws.device_type,
  count(distinct ws.website_session_id) as sessions,
  count(distinct o.order_id) as session_orders,
  round((count(distinct o.order_id)/count(distinct ws.website_session_id)) * 100,2) as conversion_rate_pct
from 
  website_sessions as ws
left join 
  orders as o
on o.website_session_id = ws.website_session_id
where 
  ws.created_at < '2012-05-11'and 
  ws.utm_source = 'gsearch' and 
  ws.utm_campaign = 'nonbrand'
group by 1
```
