```sql
CREATE TEMPORARY TABLE first_page_view
select 
  website_session_id,
  min(website_pageview_id) as min_pv_id
 from workspace.mavenfuzzyfactory.website_pageviews
where website_pageview_id < 1000
group by 1;
---------------------------------------------------------------------------------
select 
  -- fpv.website_session_id,
   wp.pageview_url as landing_page,
   count (distinct fpv.website_session_id) as sessions_hitting_lander
from first_page_view fpv
left join website_pageviews wp
  on fpv.min_pv_id = wp.website_session_id
  and fpv.min_pv_id = wp.website_pageview_id
group by 1
order by 1


select * from first_page_view
```
