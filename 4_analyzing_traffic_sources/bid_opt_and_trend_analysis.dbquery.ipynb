```sql
select
  primary_product_id,
  -- order_id,
  -- items_purchased,
  count(distinct case when items_purchased = 1 then order_id else null end) as single_order_items,
  count(distinct case when items_purchased > 1 then order_id else null end) as two_order_items

From orders
where order_id between 31000 and 32000
group by 1
```
