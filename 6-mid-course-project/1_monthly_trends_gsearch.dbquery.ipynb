```sql
use workspace.mavenfuzzyfactory;


where utm_source = 'gsearch'


select 
  utm_source,
  utm_campaign
from website_sessions 
group by 1,2;

select * from website_sessions limit 10;
select * from website_pageviews limit 10;
select * from orders limit 10; 

SELECT 
    DATE_TRUNC('month', ws.created_at) AS session_month,
    count(distinct ws.website_session_id) as sessions,
    count(distinct o.order_id) as orders
FROM website_sessions as ws
left join orders as o 
on ws.website_session_id = o.website_session_id
where ws.utm_source = 'gsearch' and ws.created_at <= '2012-11-27'
GROUP BY session_month
ORDER BY session_month
```
